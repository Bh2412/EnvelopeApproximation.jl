```julia
using Plots
using EnvelopeApproximation.GravitationalPotentials.SecondOrderODESolver
```

```julia
frequency = 2 / 5
times = LinRange(0.,  / 2, 1_000)
s_vals = -(frequency ^ 2) * cos.(frequency * times)
s = Source(times, s_vals)
numerical_solution = ode_solution(s)
exact_solution = cos.(frequency * times) .- 1
using Plots
plot(times, numerical_solution, label="numeric")
plot!(times, exact_solution, label="analytic")
```

```julia
plot(times, (numerical_solution - exact_solution))
```
